```python
CAT_NO = "6345.0" #WPI Cat

#Series Map of employmrnt rate
"""
series_map = {
    "A2603609J": "Quarter_index_Total_Ourly_Rate_Pay_trend",
    "A2713849C": "Quarter_index_Total_Ourly_Rate_Pay_adjusted",
    "A2603609J": "Quarter_index_Total_Ourly_Rate_Pay_Original"
}
"""
START_YEAR = "2000"

RAW_DIR = Path('Data/raw')
```

```python
def download_wpi_data(cat_no, series_id, label):
    #############################################################
    # Input : Cat No, Series_id, and Label
    # Output : Return : Tuple(df Monthly, df quarterly), File on data/raw/ Directory
    ##############################################################

    
    print(f"\nProcessing {label}...")
    
    # Fetch data tuple with readabs
    df, meta = ra.read_abs_series(cat=cat_no, series_id=series_id)
    
    # Ensure it is a DataFrame and filter starting years
    if isinstance(df, pd.Series):
        df = df.to_frame(name='value')
    
    df_filtered = df[df.index >= START_YEAR].copy()
    
    # Process Monthly Data
    # Rename Index and Column to the label
    df_filtered_final = df_filtered.rename_axis('date').rename(columns={df.columns[0]: label})
    
    data_path = RAW_DIR / f"abs_{label}_.csv"
    df_filtered_final.to_csv(data_path)
    print(f"Saved: {data_path}")

    return df_filtered_final
```

```python
Season_adj_wpi = download_wpi_data(CAT_NO, 'A2713849C', 'Quarter_index_Total_Ourly_Rate_Pay_adjusted')


print("Quarter_index_Total_Ourly_Rate_Pay_adjusted")
Season_adj_wpi.tail()
```

```python
print("Percentage change (QtoQ Quarter_index_Total_Ourly_Rate_Pay_adjusted")
pc_qtoq_wpi = ra.percent_change(Season_adj_wpi, 1)
pc_qtoq_wpi
```

```python
pc_qtoq_wpi = pc_qtoq_wpi.rename_axis('date').rename(columns={pc_qtoq_wpi.columns[0]: 'QtoQ percent Change'})
    
data_path = RAW_DIR / f"abs_QtoQ percent Change_.csv"
pc_qtoq_wpi.to_csv(data_path)
print(f"Saved: {data_path}")
```
